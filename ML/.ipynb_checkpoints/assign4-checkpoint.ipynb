```python
current_x=2
rate=0.01
previous_step_size=1
precision=0.000001
max_iter=10000
iters=0
df=lambda x: 2*(x+3)
```

```python
while previous_step_size>precision and iters<max_iter:
    previous_x=current_x
    current_x=current_x-rate*df(previous_x)
    previous_step_size=abs(current_x-previous_x)
    iters=iters+1

    print("iteration :",iters,"\n X value is",current_x)

print("The local minimum occurs at", current_x)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def objective(x):
    return (x+3)**2
```

```python
x1=np.linspace(-15,15,100)
```

```python
plt.plot(x1,objective(x1))
plt.plot(current_x,objective(current_x),marker='o')
```
